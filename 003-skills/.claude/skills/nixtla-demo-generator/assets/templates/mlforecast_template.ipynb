## Configure Model with Features

```python
mlf = MLForecast(
    models=[lgb.LGBMRegressor(verbosity=-1)],
    freq='D',
    lags=[1, 7, 14],
    lag_transforms={
        1: ['expanding_mean'],
        7: ['rolling_mean_7']
    },
    date_features=['dayofweek', 'month'],
    target_transforms=[Differences([1])]
)
```
